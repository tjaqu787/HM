```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
genders =['../inputs/male_cx_data.csv','../inputs/female_cx_data.csv','../inputs/family_cx_info.csv']
```

```python
import tensorflow
```

```python
#1
data = pd.read_csv(genders[0], index_col='customer_id')
data = data.fillna(0)
train,test = train_test_split(data,train_size=20000)
del test
del data
distances_agglomerative = np.arange(30,130,3)
distances_agglomerative = distances_agglomerative.reshape((len(distances_agglomerative), 1))
distances_agglomerative = np.append(distances_agglomerative, np.zeros((len(distances_agglomerative), 5)), axis=1)
for i in range(distances_agglomerative.shape[0]):
    print(i)
    agglom = AgglomerativeClustering(distance_threshold=None,n_clusters=int(distances_agglomerative[i,0]))
    agglom.fit(train)
    try:
        distances_agglomerative[i, 1] = (metrics.davies_bouldin_score(train, agglom.labels_))
        distances_agglomerative[i, 2] = (metrics.silhouette_score(train, labels=agglom.labels_))
        distances_agglomerative[i, 3] = (metrics.calinski_harabasz_score(train, agglom.labels_))
        labels = agglom.labels_
        distances_agglomerative[i, 4] = len(set(labels)) - (1 if -1 in labels else 0)
        distances_agglomerative[i, 5] = list(labels).count(-1)
    except:
        print(f'No convergence on {distances_agglomerative[i, 0]}')
agglom_plot = pd.DataFrame(distances_agglomerative,
                          columns=['index', 'davies_bouldin_score', 'silhouette_score', 'calinski_harabasz_score',
                                   'n_features', 'n_noise']).set_index('index')
agglom_plot[['davies_bouldin_score', 'silhouette_score']].plot()
agglom_plot.to_csv('../Models/clustering_outputs/agglom_mixed.csv')
```
